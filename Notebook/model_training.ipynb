```python
df.drop(columns=["day","month","year"],axis=1,inplace=True)
```

```python
df["Classes"]=np.where(df["Classes"].str.contains("not fire"),0,1)
```

```python
X=df.drop("FWI",axis=1)
Y=df["FWI"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

# feature _selection based on correlation

```python
Heat=X_train.corr()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(Heat,annot=True)
plt.show()
```

```python
def correlation(dataset,threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
         for j in range(i):
             if abs(corr_matrix.iloc[i,j]>threshold):
                 col_name=corr_matrix.columns[i]
                 col_corr.add(col_name)
    return col_corr
  
    
    
```

```python
drop_features=correlation(X_train,0.85)
```

# dropping features having corr=85%

```python
X_train.drop(drop_features,axis=1,inplace=True)
X_test.drop(drop_features,axis=1,inplace=True)
X_train.shape,X_test.shape
```

## feature scaling or standardisation



```python
from sklearn.preprocessing import StandardScaler
Scaler=StandardScaler()
X_train_scaler=Scaler.fit_transform(X_train)
X_test_scaler=Scaler.transform(X_test)
```

```python
X_train_scaler
```

```python
fig,ax=plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train_scaler)
plt.title("after scaling")
plt.subplot(1,2,2)
sns.boxplot(data=X_train)
plt.title("before sacling")
plt.tight_layout()
plt.show()
```

# Model training

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
l_reg=LinearRegression()
l_reg.fit(X_train_scaler,Y_train)
y_predict=l_reg.predict(X_test_scaler)
score=r2_score(Y_test,y_predict)
print(score)
plt.scatter(y_predict,Y_test)
plt.show()
```

# lasso_model

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score
lasso=Lasso()
lasso.fit(X_train_scaler,Y_train)
y_predict=lasso.predict(X_test_scaler)
score=r2_score(Y_test,y_predict)
print(score)
plt.scatter(y_predict,Y_test)
plt.show()
```

# ridge_model

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score
ridge=Ridge()
ridge.fit(X_train_scaler,Y_train)
y_predict=ridge.predict(X_test_scaler)
score=r2_score(Y_test,y_predict)
print(score)
plt.scatter(y_predict,Y_test)
plt.show()
```

# elasticnet_model

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import r2_score
elastic=ElasticNet()
elastic.fit(X_train_scaler,Y_train)
y_predict=elastic.predict(X_test_scaler)
score=r2_score(Y_test,y_predict)
print(score)
plt.scatter(y_predict,Y_test)
plt.show()
```

```python
from sklearn.linear_model import LassoCV
lassocv=LassoCV(cv=5)
lassocv.fit(X_train_scaler,Y_train)
l_predict=lassocv.predict(X_test_scaler)
```

```python
plt.scatter(Y_test,l_predict)
plt.show()
score=r2_score(Y_test,l_predict)
print(score)
```
